```python
import os
import cv2
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle,Circle
import numpy as np
from mtcnn.mtcnn import MTCNN
import tensorflow as tf
from keras.preprocessing import image
import matplotlib.image as mpimg
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from keras import Model
import pandas as pd
```

```python
base_dir  = 'archive (3)/test/children'
# os.chdir(base_dir)
df=os.listdir("archive (3)/test/children")
test_adults=[]
for i in range(len(df)):
    test_adults.append(os.path.join(base_dir,df[i]))
n=0
test_adults
```

```python
for i in test_adults:
    detector = MTCNN()
    img = cv2.imread(i)
    width = int(img.shape[1] * 150 / 100)
    height = int(img.shape[0] * 150 / 100)
    img=cv2.resize(img, (width,height), interpolation = cv2.INTER_AREA)
    img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    faces = detector.detect_faces(img)
    if faces ==[]:
        print("DROPED image")
        continue
    print(faces)

    
    for i in range(len(faces)):
        
        if  faces[i]['confidence']<0.9 :
            continue
        x1, y1, width, height = faces[i]['box']
        x2, y2 = x1 + width, y1 + height
        # define subplot
        plt.subplot(1, len(faces), i+1)
        plt.axis('off')
        # plot face
        plt.imshow(img[y1:y2, x1:x2])
#         print(img[y1:y2, x1:x2])
        standart_size =cv2.resize(img[y1:y2, x1:x2], (350,350), interpolation = cv2.INTER_AREA)
        cv2.imwrite('face_archive/test/children/{}.png'.format(n), cv2.cvtColor(standart_size, cv2.COLOR_BGR2RGB)) 
        n+=1
        
#     for face in faces:
#         # get coordinates
#         x, y, width, height = face['box']
#         # create the shape
#         rect = Rectangle((x, y), width, height, fill=False, color='red')
#         # draw the box
#         ax.add_patch(rect)
#         # draw the dots
#         for key, value in face['keypoints'].items():
#             # create and draw dot
#             dot = Circle(value, radius=2, color='red')
#             ax.add_patch(dot)
    plt.show()
    
```
